Given 3D vectors [40,120,60] and [60,50,90]

### Q1 draw out these vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Initialize the two vectors
colour1 = [40, 120, 60]
colour2 = [60, 50, 90]

ax = plt.figure().add_subplot(projection='3d')

# Draw out the vectors
ax.quiver(
    0, 0, 0,
    colour1[0], colour1[1], colour1[2],
    color='blue', linewidth=2, label='Colour 1'
)
ax.quiver(
    0, 0, 0,
    colour2[0], colour2[1], colour2[2],
    color='red', linewidth=2, label='Colour 2'
)

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```

### Q2 Calculate the euclidian, manhattan, cosine, and dot product distances between these vectors

### Manhattan distance

```python
L1 = sum(abs(c1 - c2) for c1, c2 in zip(colour1, colour2))

print(L1)
```

```python
#An alternate approach

a = np.array(colour1)
b = np.array(colour2)

np.linalg.norm((a - b), ord=1)
```

### Euclidian distance

```python
import math
L2 = math.sqrt(sum(math.pow(colour1[i] - colour2[i], 2) for i in range(len(colour1))))

print(L2)
```

```python
np.linalg.norm((a - b), ord=2)
```

### Cosine distance

```python
dot = sum(c1*c2 for c1, c2 in zip(colour1, colour2))
magnitude_colour1 = math.sqrt(sum(c**2 for c in colour1))
magnitude_colour2 = math.sqrt(sum(c**2 for c in colour2))
cos_similarity = dot/(magnitude_colour1*magnitude_colour2)
print(cos_similarity)
cosine = 1 - cos_similarity

print(cosine)
```

### Dot product

```python
dot = sum(c1*c2 for c1, c2 in zip(colour1, colour2))

print(dot)
```

```python
#alternate method

np.dot(a,b)
```

### Q3 Can you come up with a colour that has a smaller cosine distance from the first vector?

```python
colour1 = [40,120,60]
colour2 = [60,50,90]

colour3 = [abs(c1 + c2)/2 for c1, c2 in zip(colour1, colour2)]
print(colour3)
```

```python
ax = plt.figure().add_subplot(projection='3d')

# Plot a sin curve using the x and y axes.
ax.scatter(xs=colour1[0], ys=colour1[1], zs=colour1[2], zdir='z', label='Colour1')

ax.scatter(xs=colour2[0], ys=colour2[1], zs=colour2[2], zdir='z', label='Colour2')
ax.scatter(xs=colour3[0], ys=colour3[1], zs=colour3[2], zdir='z', label='NewColour')

ax.scatter(0, 0, 0, zdir='z', label='Origin', c='black')

ax.quiver(0,0,0,colour1[0],colour1[1],colour1[2], color='black', arrow_length_ratio=0.1)
ax.quiver(0,0,0,colour2[0],colour2[1],colour2[2], color='black', arrow_length_ratio=0.1)

#Add vector 3 to the diagram

ax.quiver(0,0,0,colour3[0],colour3[1],colour3[2], color='black', arrow_length_ratio=0.1)

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 150)
ax.set_ylim(0, 150)
ax.set_zlim(0, 150)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35, roll=0)

plt.show()
```
